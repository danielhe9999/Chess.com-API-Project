```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests

df = pd.read_csv('Draft.csv')
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
country = df['Country'].tolist()
```

```python
shortenedCountry = []

for item in country:
    string = item
    splitted = string[34:36]
    shortenedCountry.append(splitted) 

df['Country'] = shortenedCountry
```

```python
def getplayerinfo(username, timecontrol, last_or_record):
    link = "https://api.chess.com/pub/player/" + username + "/stats"
    desiredinfo = requests.get(link).json().get(timecontrol).get(last_or_record)
    return desiredinfo
```

```python
#Blitz record
recordblitzlist = []
for x in range(len(listofplayers)):
        recordblitzlist.append("")
        
for player in listofplayers:
    try:     
        recordblitz = getplayerinfo(player, 'chess_blitz', 'record')
        recordblitzlist[listofplayers.index(player)] = recordblitz
    except:
        recordblitzlist[listofplayers.index(player)] = "N/A"
```

```python
#Bullet record
recordbulletlist = []
for x in range(len(listofplayers)):
        recordbulletlist.append("")
        
for player in listofplayers:
    try:     
        recordbullet = getplayerinfo(player, 'chess_bullet', 'record')
        recordbulletlist[listofplayers.index(player)] = recordbullet
    except:
        recordbulletlist[listofplayers.index(player)] = "N/A"
```

```python
#Rapid record
recordrapidlist = []
for x in range(len(listofplayers)):
        recordrapidlist.append("")
        
for player in listofplayers:
    try:     
        recordrapid = getplayerinfo(player, 'chess_rapid', 'record')
        recordrapidlist[listofplayers.index(player)] = recordrapid
    except:
        recordrapidlist[listofplayers.index(player)] = "N/A"
```

```python
#Daily record
recorddailylist = []
for x in range(len(listofplayers)):
        recorddailylist.append("")
        
for player in listofplayers:
    try:     
        recorddaily = getplayerinfo(player, 'chess_daily', 'record')
        recorddailylist[listofplayers.index(player)] = recorddaily
    except:
        recorddailylist[listofplayers.index(player)] = "N/A"
```
